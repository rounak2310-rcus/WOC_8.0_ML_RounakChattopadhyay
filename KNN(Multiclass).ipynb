```python
class Modify():
  def __init__(self):
    pass

  def fit(self, X):
    self.min = X.min(axis=0)
    self.max = X.max(axis=0)
    self.mean = X.mean(axis=0)
    self.err = 1e-8
    self.std = X.std(axis=0) + self.err
    self.range = self.max-self.min + self.err

  def norm(self, X):
    try:
      return (X-self.min)/self.range
    except Exception as error_type:
      print("Error: ", error_type)

  def z_score(self, X):
    try:
      return (X-self.mean)/self.std
    except Exception as error_type:
      print("Error: ", error_type)
```

```python
class KNN():
  def __init__(self, K = 4):
    self.K = K

  def fit(self, X, Y):
    self.X = np.array(X)
    self.Y = np.array(Y)
    self.modify = Modify()
    self.modify.fit(self.X)
    self.A = self.modify.z_score(self.X)


  def predict(self, P):
    B = self.modify.z_score(np.array(P)).reshape(1, -1)
    if (B.shape[1]==self.A.shape[1]):
      C = abs(self.A-B).sum(axis=1)
      print(C.shape)
      idx = np.argsort(C)[:self.K]
      U, V = np.unique(self.Y[idx], return_counts=True)
      return U[np.argmax(V)]
    else:
      B = B.reshape(-1, self.A.shape[1])
      G = []
      for r in range(B.shape[0]):
        C = abs(self.A-B[r, :]).sum(axis=1)
        idx = np.argsort(C)[:self.K]
        U, V = np.unique(self.Y[idx], return_counts=True)
        G.append(U[np.argmax(V)])
      return G
```

```python
model = KNN(K=35)
model.fit(X_train, y_train)
```

```python
y_validation_prediction = model.predict(X_validation)
```

```python
(y_validation_prediction==y_validation).sum()/len(y_validation)
```

```python
classes = np.unique(y_validation)
f1s = []

for c in classes:
    tp = np.sum((y_validation_prediction == c) & (y_validation == c))
    fp = np.sum((y_validation_prediction == c) & (y_validation != c))
    fn = np.sum((y_validation_prediction != c) & (y_validation == c))

    if tp == 0:
        f1 = 0.0
    else:
        precision = tp / (tp + fp)
        recall = tp / (tp + fn)
        f1 = 2 * precision * recall / (precision + recall)

    f1s.append(f1)

macro_f1 = np.mean(f1s)
print("Macro F1 score:", macro_f1)
```
